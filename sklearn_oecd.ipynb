```python
import pandas as pd
oecd_bli = pd.read_csv("oecd_bli_2015.csv")
oecd_bli
```

```python
gdp_per_capita = pd.read_csv("gdp_per_capita.csv", encoding='latin1', delimiter='\t', thousands=',')
gdp_per_capita
```

```python
gdp_per_capita.info()
```

```python
oecd_bli_total = oecd_bli[oecd_bli['INEQUALITY'] == 'TOT']
oecd_bli_total
```

```python
oecd_bli_total_pivot = oecd_bli_total.pivot(index="Country", columns='Indicator', values="Value")
oecd_bli_total_pivot
```

```python
oecd_bli_total_pivot_life_satisfaction = oecd_bli_total_pivot['Life satisfaction']
oecd_bli_total_pivot_life_satisfaction
```
